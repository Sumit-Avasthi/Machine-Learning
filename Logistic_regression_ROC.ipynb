```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_classes=2,random_state=42)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
# Creating a dummy model with 0 as default output
dummy_model_prob = [0 for _ in range(len(y_test))]
dummy_model_prob
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
y_pred_prob = model.predict_proba(X_test)
```

```python
y_pred_prob
```

```python
model_proba = y_pred_prob[:,1]
model_proba
```

```python
dummy_model_roc = roc_auc_score(y_test,dummy_model_prob)
model_proba_roc = roc_auc_score(y_test,model_proba)
print(dummy_model_roc)
print(model_proba_roc)
```

```python
dummy_fpr,dummy_tpr,_ = roc_curve(y_test,dummy_model_prob)
model_fpr,model_tpr,threshold= roc_curve(y_test,model_proba)
```

```python
import matplotlib.pyplot as plt

plt.plot(dummy_fpr,dummy_tpr,linestyle='--',label='Dummy Model')
plt.plot(model_fpr,model_tpr,marker='.',label='Real Model')
plt.legend()
plt.show()
```
